```python
# Custom Exception
class LoginError(Exception):


    try:
            
        correct_password = "1234"
        attempts = 3
        
        while attempts > 0:
            pwd = input("Enter password: ")
            
            if pwd == correct_password:
                print("Login successful!")
                break
            else:
                attempts -= 1
                print("Wrong password! Attempts left:", attempts)
    
        if attempts == 0:
            raise LoginError("Too many failed attempts!")
    
    except LoginError as e:
        print(e)
    except ValueError:
        print("Invalid Input!")
```

```python
#convert string to float with exception handling
try:
    value = float(input("Enter a number: "))
    print("Float value:", value)
except ValueError:
    print("Invalid input! Not a float.")
```

```python
#atm withdrawal with insufficient balance handling
balance = 5000

try:
    amount = int(input("Enter withdrawal amount: "))
    if amount > balance:
        raise Exception("Insufficient balance!")
    else:
        balance -= amount
        print("Withdrawn:", amount)
        print("Remaining balance:", balance)
except ValueError:
    print("Invalid amount!")
except Exception as e:
    print(e)
```

```python
#validate age input with custom exception
class AgeError(Exception):
    try:
        age = int(input("Enter age: "))
        if age < 18:
            raise AgeError("Age must be 18 or above!")
        print("Valid age")
    except ValueError:
        print("Invalid input!")
    except AgeError as e:
        print(e)
```

```python
#handle multiple exception 
try:
    a = int(input("Enter number: "))
    b = int(input("Enter another number: "))
    c=input("Enter a text:")
    result = a / b
    
    lst = [1, 2, 3]
    print(lst[a])
    r=c + result
    print(r)

except TypeError:
    print("Type error occurred! Cannot mix different data types.")
except ValueError:
    print("Invalid input!")
except ZeroDivisionError:
    print("Cannot divide by zero!")
except IndexError:
    print("Index error!")
```

```python
#use finally
try:
    f = open("sample.txt", "r")
    print(f.read())
except FileNotFoundError:
    print("File not found!")
finally:
    try:
        f.close()
        print("File closed.")
    except:
        print("No file to close.")
```

```python
#Remove duplicate lines.
with open("sam.txt", 'w') as file:
    file.write("Harsitha\nAnu\nRahul\nAnu\nMegha\n")

with open("sam.txt", 'r') as file:
    lines = file.readlines()

unique = set(lines)

with open("sam.txt", 'w') as file:
    file.writelines(unique)

with open("sam.txt", 'r') as file:
    print(file.read())
```

```python
#Store and calculate student marks
with open("marks.txt", 'w') as file:
    file.write("Harsitha 80 90 85\n")
    file.write("Anu 70 88 92\n")
    file.write("Rahul 60 75 80\n")

with open("marks.txt", 'r') as file:

    for line in file:
        data = line.split()

        name = data[0]
        mark1 = int(data[1])
        mark2 = int(data[2])
        mark3 = int(data[3])

        total = mark1 + mark2 + mark3
        average = total / 3

        print("Name:", name)
        print("Total:", total)
        print(f"Average:  {average:.2f}")
        print()
```

```python
#Parse CSV-like file manually
with open("data.txt", 'w') as file:
    file.write("Harsitha,80,90,85\n")
    file.write("Anu,70,88,92\n")
    file.write("Rahul,60,75,80\n")

with open("data.txt", 'r') as file:

    for line in file:
        name, mark1, mark2, mark3 = line.strip().split(',')

        total = int(mark1) + int(mark2) + int(mark3)

        print("Name:", name)
        print("Mark1:", mark1)
        print("Mark2:", mark2)
        print("Mark3:", mark3)
        print("Total:", total)
        print()
```

```python
#Update records inside file
with open("student.txt", 'w') as file:
    file.write("Harsitha Chennai\n")
    file.write("Anu Kerala\n")
    file.write("Rahul Mumbai\n")

with open("student.txt", 'r') as file:
    data = file.readlines()

new_data = []

for line in data:

    if "Rahul" in line:
        new_data.append("Rahul Delhi\n")

    else:
        new_data.append(line)

with open("student.txt", 'w') as file:
    file.writelines(new_data)

with open("student.txt", 'r') as file:
    print(file.read())
```
