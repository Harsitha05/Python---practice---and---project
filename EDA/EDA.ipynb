```python
#titanic dataset

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("titanic.csv")

print(df.head())

print(df.isnull().sum())

df['Age'] = df['Age'].fillna(df['Age'].median())

df['Fare'] = df['Fare'].fillna(df['Fare'].median())

df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])

df.drop_duplicates(inplace=True)

if 'Cabin' in df.columns:
    df.drop('Cabin', axis=1, inplace=True)

print("Data cleaned successfully")

age = np.array(df['Age'])

print("Average Age:", np.mean(age))
print("Maximum Age:", np.max(age))
print("Minimum Age:", np.min(age))

plt.figure(figsize=(5,4))
df['Survived'].value_counts().plot(kind='bar')
plt.title('Survival Count')
plt.xlabel('Survived')
plt.ylabel('Count')
plt.show()

plt.figure(figsize=(5,4))
df['Age'].plot(kind='hist', bins=20)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

plt.figure(figsize=(5,5))
df['Pclass'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Passenger Class Percentage')
plt.ylabel('')
plt.show()

plt.figure(figsize=(6,4))
plt.plot(df['Fare'].head(50))
plt.title('Fare Line Chart')
plt.xlabel('Passengers')
plt.ylabel('Fare')
plt.show()

plt.figure(figsize=(6,4))
plt.scatter(df['Age'], df['Fare'])
plt.title('Age vs Fare')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()

plt.figure(figsize=(5,4))
plt.boxplot(df['Age'])
plt.title('Box Plot of Age')
plt.show()

df.to_csv("cleaned_titanic.csv", index=False)
```
